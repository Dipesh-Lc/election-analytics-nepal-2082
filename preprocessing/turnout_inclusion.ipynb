```python
voter82 = pd.read_csv("data/raw/All_Nepal_district_level_data_voter_data_2082.csv")
```

```python
voter82.head()
```

```python
voter82["age_70+"] = voter82["age_70-79"] + voter82["age_80-89"] + voter82["age_90-99"] + voter82["age_100+"]
```

```python
voter82["female_ratio"] = voter82["gender_महिला"] / voter82["total_voters"]
```

```python
voter82["age_18-29_ratio"] = voter82["age_18-29"] / voter82["total_voters"]
voter82["age_30-39_ratio"] = voter82["age_30-39"] / voter82["total_voters"]
voter82["age_40-49_ratio"] = voter82["age_40-49"] / voter82["total_voters"]
voter82["age_50-59_ratio"] = voter82["age_50-59"] / voter82["total_voters"]
voter82["age_60-69_ratio"] = voter82["age_60-69"] / voter82["total_voters"]
voter82["age_70+_ratio"] = voter82["age_70+"] / voter82["total_voters"]
```

```python
temp = voter82[['district_id','female_ratio', 'age_18-29_ratio', 'age_30-39_ratio', 'age_40-49_ratio', 'age_50-59_ratio', 'age_60-69_ratio', 'age_70+_ratio',"total_voters"]]
```

```python
temp.info()
```

```python
base79.drop(columns=['female_ratio'], inplace=True)
base79 = base79.merge(temp, on='district_id', how='left')
```

```python
base79.drop(columns=['female_ratio'], inplace=True)
```

```python
base79.head()
```

```python
stat79 = pd.read_csv("data/raw/district_wise_result_statistics_2079.csv")
```

```python
stat79["female_ratio"] = stat79["Total_female_voters"] / stat79["Total_eligible_voters"]
```

```python
base79 = base79.merge(stat79[['district_id', 'female_ratio','Turnout_Rate_PR']], on='district_id', how='left')
```

```python
base79.rename(columns={'Turnout_Rate_PR': 'valid_turnout_79'}, inplace=True)
```

```python
base79["valid_turnout_74"] = base79["valid_turnout_74"].round(2)
base79.head()
```

```python
stat74 = pd.read_csv("data/raw/district_wise_pr_result_statistics_2074.csv")
```

```python
stat74.head()
```

```python
stat74["valid_turnout_74"] = (stat74["Total_valid_pr_votes"] / stat74["total_voters"])*100
```

```python
base79 = base79.merge(stat74[['district_id', 'valid_turnout_74']], on='district_id', how='left')
```

```python
base79.info()
```

```python
base79.rename(columns={'valid_turnout_79': 'valid_turnout','valid_turnout_74': 'valid_turnout_lag'}, inplace=True)
```

```python
import numpy as np

base79["facebook_log"] = np.log1p(
    base79["Facebook_Presence (In Thousands)"]
)
base79["leader_following_log"] = np.log1p(
    base79["Top_leader_fb (in Thousands)"]
)
```

```python
base79.to_csv("data/processed/training_pr_base.csv", index=False)
```
