```python
import pandas as pd

# Load party database
party_df_74 = pd.read_csv("data/processed/pr_votes_long_2074.csv")
party_df_79 = pd.read_csv("data/processed/pr_votes_long_2079.csv")
```

```python
party_df_74.head(10)
```

```python
RPP_lag79 = party_df_74[party_df_74['party_id'].isin(["P5", "P96", "P97"])].groupby('district_id').agg({'Vote_Share': 'sum'}).reset_index()
```

```python
RPP_lag79.rename(columns={'Vote_Share': 'vote_share_lag'}, inplace=True)
```

```python
RPP_lag79["party_id"] = "P005"
```

```python
RPP_lag79.head()
```

```python
RSP_lag82 = party_df_79[party_df_79['party_id'].isin(["P4", "P92"])].groupby('district_id').agg({'Vote_Share': 'sum'}).reset_index()
```

```python
RSP_lag82["party_id"] = "P004"
```

```python
RSP_lag82.rename(columns={'Vote_Share': 'vote_share_lag'}, inplace=True)
```

```python
RSP_lag82.tail()
```

```python
NCP_lag82 = party_df_79[party_df_79['party_id'].isin(["P3", "P8", "P94"])].groupby('district_id').agg({'Vote_Share': 'sum'}).reset_index()
```

```python
NCP_lag82["party_id"] = "P003"
```

```python
NCP_lag82.rename(columns={'Vote_Share': 'vote_share_lag'}, inplace=True)
```

```python
NCP_lag82.tail()
```

```python
model_base79 = pd.read_csv("data/processed/training_pr_base.csv")
```

```python
model_base79.head()
```

```python
model_base79.head()
```

```python
model_base79 = model_base79.merge(
    RPP_lag79[['district_id', 'party_id', 'vote_share_lag']],
    on=['district_id', 'party_id'],
    how='left',
    suffixes=('', '_new')
)

model_base79['vote_share_lag'] = (
    model_base79['vote_share_lag_new']
    .combine_first(model_base79['vote_share_lag'])
)

model_base79.drop(columns='vote_share_lag_new', inplace=True)

```

```python
model_base79["vote_share_change"] = model_base79["vote_share"] - model_base79["vote_share_lag"]
```

```python
model_base79.to_csv("data/processed/training_pr_base.csv", index=False)
```
